# 1. git --version

-> Shows the current git bash version

```
$ git --version

  git version 2.4.5.1 windows.1
```

---


# 2. git config  --global username "username"

# git config  --global email "email"

-> Sets "Username" and "Email" in Git Bash on global scope.

---

# 3. git init

```
$ git init

  Initialized empty Git repository in /path
```

---

# 4. git status

-> Shows the current status of local Git Repository.

```
$ git status

  On branch Master
  No Commits yet

  Untracked Files:
        index.html
```

---

# 5. git add .

-> Adds all untracked files to staging area.

---

# 6. git status

```
$ git status

  On branch Master
  No Commits yet

  Changes to be committed:
      new file: index.html
```

---

# 7. git commit -m "commit message"

-> Commit the changes with given commit message

```
$ git commit -m "test commit"

  1 file changed, 1 insertion(+)
```

---

# 8. git log --oneline

```
    11aea52 (HEAD -> master) test commit
```

---

# 9. git remote add origin "repository URL"

-> Add a remote GitHub repository as origin.

```
  $ git remote add origin https://github.com/username/repository.git
```
---

# 10. git push -u origin master

```
$ git push -u origin master

  Enumerating objects: 1, done.
  Counting Objects: 1, done.
  Writing Objects:100% done.

  Branch 'master' setup to track remote

  branch 'master' from 'origin'.
  
```
